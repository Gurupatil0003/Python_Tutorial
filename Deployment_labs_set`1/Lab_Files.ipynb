```python
print("Temperature Converter")
print("---------------------")

choice = input("Choose an option:\n1. Fahrenheit to Celsius\n2. Celsius to Fahrenheit\nEnter 1 or 2: ")

if choice == '1':
    f = float(input("Enter temperature in Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"{f} degrees Fahrenheit is equal to {c:.2f} degrees Celsius.")

elif choice == '2':
    c = float(input("Enter temperature in Celsius: "))
    f = (c * 9/5) + 32
    print(f"{c} degrees Celsius is equal to {f:.2f} degrees Fahrenheit.")

else:
    print("Invalid choice. Please enter 1 or 2.")
```

**A program that calculates the area and perimeter of a rectangle.**

```python
print("Rectangle Calculator")
print("---------------------")

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

area = length * width
perimeter = 2 * (length + width)

print(f"The area of the rectangle is: {area:.2f} square units")
print(f"The perimeter of the rectangle is: {perimeter:.2f} units")
```

# **3. A program that generates a random password of a specified length.**




```python
import random
import string

print("Random Password Generator")
print("-------------------------")

password_length = int(input("Enter the desired length for the password: "))

if password_length > 0:
    password = ''.join(random.choice(string.ascii_letters + string.digits + string.punctuation) for _ in range(password_length))
    print(f"Generated Password: {password}")
else:
    print("Invalid password length. Please enter a positive integer.")
```

# **4. A program that calculates the average of a list of numbers**

```python
print("Average Calculator")
print("-------------------")

numbers = input("Enter a list of numbers separated by spaces: ")
numbers_list = [float(num) for num in numbers.split()]#Enter a list of numbers separated by spaces: 34 4
                                                               #[34.0, 4.0]
if numbers_list:
    average = sum(numbers_list) / len(numbers_list)
    print(f"The average is: {average:.2f}")
else:
    print("No numbers provided. Please enter a list of numbers.")
```
