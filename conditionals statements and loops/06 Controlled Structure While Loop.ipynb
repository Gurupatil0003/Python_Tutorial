```python
count = 0

while count < 5:
    print(count)
    count = count + 1
```

```python
n = int(input("Enter number N :"))
sum = 0
temp = n
while n > 0 :
  sum = sum + n
  n = n - 1
print("Sum of",temp,"numbers is:",sum)
```

```python
# product of n numbers using while loop
n = int(input("enter the number: "))
prod = 1
temp = n
while n>0:
    prod = prod*n
    n = n-1
print("Product of",temp,'numbers is : ', prod)
```

```python
# While Loop Example

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while year != 1973:
    year = dates[i]
    i = i + 1
    print(year)

print("It took ", i ,"repetitions to get out of loop.")
```

```python
inc_count = 0
dec_count = 10

while inc_count > 5 or dec_count > 5:
    print(inc_count, dec_count)
    
    inc_count = inc_count + 1
    dec_count = dec_count - 2

print('Final count values ', inc_count, dec_count) 
```

# While loop with else


```python
# Python program to demonstrate
# while-else loop

i = 0
while i < 4:
	i += 1
	print(i)
else: # Executed because no break in for
	print("No Break\n")

```

```python
i = 0
while i < 4:
	i += 1
	print(i)
	break
else: # Not executed as there is a break
	print("No Break")
```
